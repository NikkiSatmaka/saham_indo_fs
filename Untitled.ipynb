```python
import pandas as pd
import requests
import os
```

```python
xls_loc = 'lap_keuangan/XLSX'

bs_indo = [file for file in os.listdir(xls_loc) if file[5:7] == 'BS']
cs_indo = [file for file in os.listdir(xls_loc) if file[5:7] == 'CS']
is_indo = [file for file in os.listdir(xls_loc) if file[5:7] == 'IS']
```

```python
"""
Create a dictionary where each item is a panda series of the financial statement
"""

data_bs = {}
for file in bs_indo:
    data_bs[file[:4]] = pd.read_excel(f'{xls_loc}/{file}')
    
data_is = {}
for file in is_indo:
    data_is[file[:4]] = pd.read_excel(f'{xls_loc}/{file}')

data_cs = {}
for file in cs_indo:
    data_cs[file[:4]] = pd.read_excel(f'{xls_loc}/{file}')
```

```python
"""
Drop all rows and columns which only contains NA
"""

for key, item in data_bs.items():
    item.dropna(axis=0, how='all', inplace=True)
    item.dropna(axis=1, how='all', inplace=True)
    data_bs[key] = item

for key, item in data_cs.items():
    item.dropna(axis=0, how='all', inplace=True)
    item.dropna(axis=1, how='all', inplace=True)
    data_cs[key] = item
    
for key, item in data_is.items():
    item.dropna(axis=0, how='all', inplace=True)
    item.dropna(axis=1, how='all', inplace=True)
    data_is[key] = item
```
